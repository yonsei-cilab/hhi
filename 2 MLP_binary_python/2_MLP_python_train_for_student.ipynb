```
import numpy as np
import matplotlib.pyplot as plt


# ******************************************************
#                 NN forward pass
# ******************************************************

def NN(X, W1, W2, W3):
    # datanum,dim=X.shape
    # X is row-wise samples

    v1 = np.matmul(W1, X.T)
    v1 = v1.T
    Y1 = Sigmoid(v1)

    v2 = np.matmul(W2, Y1.T)
    v2 = v2.T
    Y2 = Sigmoid(v2)

    v3 = np.matmul(W3, Y2.T)
    v3 = v3.T
    Y = Sigmoid(v3)

    return Y


# ******************************************************
#                 NN backward pass
# ******************************************************
def BackpropCE(W1, W2, W3, X, D):
    alpha = 0.01
    datanum, dim = X.shape

    for k in range(datanum):
        x = np.array(X[k, :], ndmin=2).T
        d = np.array(D[k], ndmin=2)

        # NN forward
        v1 = np.matmul(W1, x)
        y1 = Sigmoid(v1)

        v2 = np.matmul(W2, y1)
        y2 = Sigmoid(v2)

        v3 = np.matmul(W3, y2)
        y = Sigmoid(v3)

        # NN Backward
        e = d - y
        delta = e

    # ***********************************
    #
    #     Fill in the blank
    #
    # ***********************************

        e1 = np.matmul(W2.T, delta2)
        delta1 = y1 * (1 - y1) * e1

        # Update
        dW1 = alpha * np.matmul(delta1, x.T)  # delta rule
        W1 = W1 + dW1

        dW2 = alpha * np.matmul(delta2, y1.T)
        W2 = W2 + dW2

        dW3 = alpha * np.matmul(delta, y2.T)
        W3 = W3 + dW3

    return W1, W2, W3


# ******************************************************
#                 NN Logistic sigmoid
# ******************************************************
def Sigmoid(x):
    # return 1.0 / (1.0 + np.exp(-x))
    return np.exp(-np.logaddexp(0, -x))

  
  
  
```

```
# ******************************************************
#                Train MLP
# ******************************************************

X=np.column_stack((X,np.ones([datanum,1])))
D = Y
outputdim=1


epochnum = 5000
ERR = np.zeros([epochnum,1])
hiddennum1 = 10
hiddennum2 = 10

W1 = 1*(2*np.random.random((hiddennum1,inputdim+1)) - 1)
W2 = 1*(2*np.random.random((hiddennum2,hiddennum1)) - 1)
W3 = 1*(2*np.random.random((outputdim,hiddennum2)) - 1)

        
for epoch in range(epochnum):
    W1, W2, W3 = BackpropCE(W1, W2, W3, X, D)
    Yhat = NN(X,W1,W2, W3)
    ERR[epoch] = 1/2*np.matmul((D-Yhat).T,(D-Yhat))/datanum
    if epoch%100 == 0:
       print("epoch = %d, Err=%f" % (epoch,ERR[epoch]))


plt.plot(ERR, 'r')
plt.show()
```

```
# ******************************************************
#                Test MLP
# ******************************************************

def plotNN2D(X,Y,W1,W2,W3):

    C1 = np.where(Y== 0)[0]
    C2 = np.where(Y== 1)[0]
    C = (C1,C2)
    colors = ("red", "green")

    for c,color in zip(C, colors):
        plt.scatter(X[c,0], X[c,1], alpha=1.0, c=color)

   
    step = 0.025
    x_axis = np.arange(0.0, 1.0+step, step)
    y_axis = np.arange(0.0, 1.0+step, step)
    X_mesh, Y_mesh = np.meshgrid(x_axis, y_axis)
    Z_mesh = np.zeros(X_mesh.shape)


    for x in range(X_mesh.shape[0]):
        for y in range(Y_mesh.shape[0]):
            input = np.array([X_mesh[x][y], Y_mesh[x][y], 1],ndmin=2)
            Z_mesh[x][y] =  NN(input,W1,W2,W3)

    plt.contour(x_axis, y_axis, Z_mesh,([0.49, 0.51]))
    plt.show()



#   ***             Plot
plotNN2D(X,Y,W1,W2,W3)
```
